```python
#mau menghilangkan \r, simbol dan angka (belum bisa)
def clean_text(text):
    text = str(text)
    text = text.lower()    
    text = re.sub('\n', " ", text)
    text = re.sub('\'', "", text)
    text = re.sub('-', " ", text)        
    text = re.sub(r"(http|https|pic.)\S+"," ",text)
    text = re.sub(r"(@|#)[:\/\w\.]+"," ",text)
    text = re.sub("[^0-9a-z]", " ", str(text))
    text = nltk.word_tokenize(text)
    text = [alay.get(a,a) for a in text]
    text = ' '.join([text for text in text])
    return text
```
